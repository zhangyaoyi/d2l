```python
import torch
import torch.nn as nn
from d2l import torch as d2l

mnist_train, mnist_test =d2l.load_data_fashion_mnist(256)
```

定义模型：2层全链接感知机
维度 784 -> 256 -> 10
最后输出 0-9的labels

```python
num_inputs, num_outputs, num_hiddens = 784, 10, 256

network = nn.Sequential(nn.Flatten(), nn.Linear(num_inputs, num_hiddens), nn.ReLU(), nn.Linear(num_hiddens, num_outputs))

def init_weights(m):
    if type(m) == nn.Linear:
        nn.init.normal_(m.weight, std=0.01)

network.apply(init_weights)

loss_function = nn.CrossEntropyLoss()

optimizer = torch.optim.SGD(network.parameters(), lr=0.1)
```

```python
epochs = 10
for epoch in range(epochs):
    train_loss = 0
    network.train()
    for X, y in mnist_train:
        pred = network(X)
        loss = loss_function(pred, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
        
    avg_loss = train_loss / len(mnist_train)
    print(f"Epoch {epoch+1}/{epochs}, Average Loss: {avg_loss:.4f}")
    
    network.eval()
    with torch.no_grad():
        for X, y in mnist_test:
            pred = network(X)
            print(y)
```
